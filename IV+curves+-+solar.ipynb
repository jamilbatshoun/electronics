```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
# define constants
mu0 = 1.257e-6
eps0 = 8.854e-12
c_light = 3e8
h=6.626e-34
hbar = 1.0546e-34
m_e = 9.1e-31
q_e = 1.6e-19
kB = 1.38e-23

# define variables
Voltage =  np.linspace(1e-3, 1, 1001)
I0 = 1e-10 # Amps
Temp = 300 # Kelvin
ISC = 4 # Amps
A= 1
Idrift = ISC - I0*(np.exp(q_e*Voltage/(A*kB*Temp))-1)
#k = omega*cmath.sqrt(mu0*eps0)*cmath.sqrt(1-sig/omega^2)

```

```python
#############################################
### begin plotting ##########################
#############################################


plt.plot(Voltage,Idrift)
plt.xlabel('Voltage (V)')
plt.ylabel('current (A)')
plt.title('solar cell IV curve')
plt.grid(True)
plt.ylim([0,ISC*1.1])
#plt.savefig("test.png")
#plt.show()
```

```python
Idrift2 = ISC - I0*(np.exp(q_e*Voltage/(1.2*kB*Temp))-1)
#k = omega*cmath.sqrt(mu0*eps0)*cmath.sqrt(1-sig/omega^2)

plt.plot(Voltage,Idrift,label="A=1")
plt.plot(Voltage,Idrift2,label="A=1.2")
plt.xlabel('Voltage (V)')
plt.ylabel('current (A)')
plt.title('solar cell IV curve')
plt.legend()
plt.grid(True)
plt.ylim([0,ISC*1.1])
#plt.savefig("test.png")
#plt.show()
```

```python
Idrift2 = ISC - I0*(np.exp(q_e*Voltage/(1.2*kB*Temp))-1)
#k = omega*cmath.sqrt(mu0*eps0)*cmath.sqrt(1-sig/omega^2)

plt.plot(Voltage,Idrift*Voltage,label="A=1")
plt.plot(Voltage,Idrift2*Voltage,label="A=1.2")
plt.xlabel('Voltage (V)')
plt.ylabel('current (A)')
plt.title('solar cell IV curve')
plt.legend()
plt.grid(True)
plt.ylim([0,ISC*1.1])
#plt.savefig("test.png")
#plt.show()
```

I'm seeing that a higher ideality factor should translate to a higher peak power...

```python
Idrift3 = ISC - I0*1*(np.exp(q_e*Voltage/(1.0*kB*Temp))-1)
Idrift4 = ISC - I0*2*(np.exp(q_e*Voltage/(1.0*kB*Temp))-1)
#k = omega*cmath.sqrt(mu0*eps0)*cmath.sqrt(1-sig/omega^2)

plt.plot(Voltage,Idrift3*Voltage,label="I0")
plt.plot(Voltage,Idrift4*Voltage,label="I0*2")
plt.xlabel('Voltage (V)')
plt.ylabel('current (A)')
plt.title('solar cell IV curve')
plt.legend()
plt.grid(True)
plt.ylim([0,ISC*1.1])
#plt.savefig("test.png")
#plt.show()
```
